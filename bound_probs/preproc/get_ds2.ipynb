```python
import pandas as pd

import glob
import re
import os
import os.path as op
```

```python
source_dir = op.join(os.environ['HOME'], 'Downloads', 'probDistortion_Sess05-Dec-2023')
reg = re.compile('.*/(?P<subject_id>[0-9]{2})/participant[0-9]+.[0-9]+_savedValues.csv')

fns = glob.glob(op.join(source_dir, '*', '*.csv'))

# [reg.match(fn).groupdict() for fn in fns]]


fns = [fn for fn in fns if reg.match(fn)]
subject_ids = [reg.match(fn).group(1) for fn in fns]
df = pd.concat([pd.read_csv(fn, index_col=[0,1,2]) for fn in fns], keys=subject_ids, names=['subject'])

df.to_csv(op.abspath('../data/ds2.csv'))
```
